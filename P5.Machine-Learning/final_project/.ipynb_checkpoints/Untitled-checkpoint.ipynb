```python
import sys
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
features_list = ['poi','salary'] # You will need to use more features
ignore_list = ['email_address'] # Ignore text based fields

### Load the dictionary containing the dataset
data_dict = pickle.load(open("final_project_dataset.pkl", "r"))
my_dataset = data_dict
# Iterate through the possible fields so we have them all:
for person in my_dataset:
    for field in my_dataset[person]:
        if field not in features_list and field not in ignore_list:
            features_list.append(field)
```

```python
df = pd.DataFrame.from_dict(my_dataset).T
```

```python
poi = {False:'blue',True:'red'}
df2 = df[['salary','total_payments','poi']].astype(float)
df2['poi'] = df2['poi'].astype(bool)


df2.plot(kind='scatter', x='salary',y='total_payments', c=df['poi'].apply(lambda x: poi[x]))
```
